**EASY PROBLEM :-** *Find the Sum of Elements in a Tuple*

```python
mytuple = (1,2,3,4,5)
sum = 0
for items in mytuple:
    sum = sum + items

print(sum)
```

**MEDIUM  PROBLEM :-** *Count Occurrence of an Element in a Tuple*

```python
mytuple = (1,2,3,4,5,6,7,8,9,1,2,3,4,5,6,7,8,9)
count = 0 

element = int(input("Enter the target value for occurence :-"))

for item in mytuple:
    if item == element:
        count+=1


print(count)
```

**HARD PROBLEM :- ** *Remove Duplicate Elements from a Tuple (Preserve Order)*

```python
mytuple = (1,2,3,4,3,2,1)
unique = []

for item in mytuple:
    if item not in unique:
        unique.append(item)

result = tuple(unique)
print(result)
```
